# Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'comments.csv'
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, order=df['Category'].value_counts().index)
plt.title('Number of Comments in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


```

```python
# Plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', hue='Sentiment', data=df, 
              palette={'Positive': '#00FF5C', 'Neutral': '#CDCDCD', 'Negative': '#FF0000'}, 
              order=df['Category'].value_counts().index)
plt.title('Sentiment Distribution in Each Category')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()



```

```python
# Filter the dataframe to include only negative sentiments
negative_df = df[df['Sentiment'] == 'Negative']

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=negative_df, 
              palette='coolwarm_r',  # Reverse the coolwarm palette
              order=negative_df['Category'].value_counts().index)
plt.title('Number of Negative Comments in Each Category')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.show()

```

```python
# Filter the dataframe to include only positive sentiments
positive_df = df[df['Sentiment'] == 'Positive']

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=positive_df, 
              palette='coolwarm', 
              order=positive_df['Category'].value_counts().index)
plt.title('Number of Positive Comments in Each Category')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.show()
```
