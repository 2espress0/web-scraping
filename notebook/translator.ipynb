```python
import pandas as pd
from googletrans import Translator

# Initialize the translator
translator = Translator()
```

```python
# Specify the file path relative to the notebook location
file_path = '../cleaned_comments.csv'

# Load the CSV file with different parameters
df = pd.read_csv(file_path, header=None)  # Try reading without headers first
df.head()  # Display the first few rows to verify the data
```

```python
# Display the DataFrame to understand its structure
print(df.head())
print(df.columns)
```

```python
# Rename the column to a standard name
df.columns = ['comments']  # Rename the first column to 'comments'

# Display the first few rows to verify
df.head()
```

```python
# Translate the comments
translated_comments = []

# Iterate through the comments and translate them
for comment in df['comments']:
    try:
        translated_comment = translator.translate(comment, src='ar', dest='en').text
        translated_comments.append(translated_comment)
    except Exception as e:
        print(f"Error translating comment: {comment}")
        print(e)
        # Skip appending the original comment if translation fails

# Create a new DataFrame with the translated comments
df_translated = pd.DataFrame(translated_comments, columns=['translated_comments'])
```

```python
# Specify the output file path relative to the notebook location
output_file_path = '../translated_comments.csv'

# Save the new DataFrame with translated comments to a new CSV file
df_translated.to_csv(output_file_path, index=False)

# Confirm the operation
df_translated.head()  # Display the first few rows to verify
```
