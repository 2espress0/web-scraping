```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sentiment analysis results from the CSV file
file_path = '../sentiment_analysis_results.csv'
df = pd.read_csv(file_path)

# Display the first few rows to verify
df.head()
```

```python
# Calculate the percentage of each sentiment category
neg_count = df['roberta_neg'].sum()
neu_count = df['roberta_neu'].sum()
pos_count = df['roberta_pos'].sum()
total_count = neg_count + neu_count + pos_count

neg_percentage = (neg_count / total_count) * 100
neu_percentage = (neu_count / total_count) * 100
pos_percentage = (pos_count / total_count) * 100

# Plot the percentage of each sentiment category
labels = ['Negative', 'Neutral', 'Positive']
sizes = [neg_percentage, neu_percentage, pos_percentage]
colors = ['red', 'royalblue', 'green']
explode = (0.1, 0, 0)  # explode the 1st slice (i.e., 'Negative')

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title(' % of Sentiment Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plot the counts of each sentiment category
plt.figure(figsize=(8, 6))
df[['roberta_neg', 'roberta_neu', 'roberta_pos']].sum().plot(kind='bar', color=['red', 'yellow', 'green'])
plt.title('Counts of Sentiment Categories')
plt.xlabel('Sentiment Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
